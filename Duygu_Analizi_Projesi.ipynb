```python

from transformers import pipeline

# İngilizce duygu analizi modeli
classifier = pipeline('sentiment-analysis')

# Kullanıcıdan veri seçimi
choice = input("Tek cümle mi analiz etmek istiyorsun? (evet/hayır): ").lower()

if choice == "evet":
    text = input("Bir cümle yaz (İngilizce): ")
    result = classifier(text)
    print(f"Cümlenin duygu durumu: {result[0]['label']} (Güven: {result[0]['score']:.2f})")
else:
    texts = input("Cümleleri aralarına ; koyarak yaz (İngilizce): ").split(";")
    results = [classifier(t.strip()) for t in texts]
    for t, r in zip(texts, results):
        print(f"'{t.strip()}' -> {r[0]['label']} ({r[0]['score']:.2f})")
```
